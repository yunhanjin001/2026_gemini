```python
import pandas as pd
import yfinance as yf
```

```python
data_path='E:\\CodeSpace\\2026_gemini\\data'
```

```python
period = "60d"
interval = "5m"   
tech_stocks = ["AAPL", "MSFT", "NVDA", "META", "GOOGL", "AMZN"]

gold = yf.download("GC=F", period=period, interval=interval)
gold['asset'] = 'GOLD'
tech_data = []
for ticker in tech_stocks:
    df = yf.download(ticker, period=period, interval=interval)
    df['asset'] = ticker
    tech_data.append(df)

tech = pd.concat(tech_data)
```

```python
#also access sp500, and other index through yfinance
tech_indexes=["^GSPC", "^DJI", "^IXIC"]
for ticker in tech_indexes:
    df = yf.download(ticker, period=period, interval=interval)
    df['asset'] = ticker
    tech_data.append(df)

tech_idx=pd.concat(tech_data)
```

```python
#get price column close
#contact all close
close=pd.concat([gold['Close'], tech['Close'], tech_idx['Close']])

mo_data_path=f'{data_path}/close_5m_3mo.csv'
close.to_csv(mo_data_path, index=False)
```
